#### Loss function: square mean error +  regulation of subunit radius
#### * xi, yi ,r :discrete

```python
dataset = scipy.io.loadmat('NN_Train.mat')
Train_rawimg = dataset['Train_img_set']
Train_target = dataset['Train_target']
print "Training images shape:",Train_rawimg.shape
print "Target exc response shape:", Train_target.shape
```

### Trainning the neural net
#### 1.1  On a toy dataset

```python
# select a toy set to test principles
ind = np.random.randint(0,Train_rawimg.shape[0],5)
XT = Train_rawimg[ind,:,:]
yT = Train_target[ind]
XT = np.reshape(XT,(XT.shape[0],-1))
print "Toy trainning set shape:", XT.shape
print "Toy garget set shape:", yT.shape
```

```python
#initialize the scratch net
D = Train_rawimg.shape[1]
num_subunits = 16
G_filter = np.ones((num_subunits,1)) # a fake Gaussian, 
#partially ignored in the gradient
sNet = ScratchNet(D,num_subunits,G_filter)
```

```python
sNet.gridInitilization('center')
print sNet.params['sub']
```

```python
sNet.printSubUnitWeights()
```

```python
# Gradient check
sNet = ScratchNet(D,num_subunits,G_filter)
sNet.gridInitilization('center')
loss, agrads = sNet.loss(XT,yT,reg = 0.0)
grad_check = sNet.numGradientCheck(XT,yT,reg = 0.0)
print grad_check
print loss
```

```python
import matplotlib.pyplot as plt
sNet = ScratchNet(D,num_subunits,G_filter)
sNet.gridInitilization('center')
stats = sNet.train(XT, yT, XT, yT,
            learning_rate=0.002, reg=1e-6,
            num_iters=1000, batch_size = 32,verbose=True)

print 'Final training loss: ', stats['loss_history'][-1]
print 'shape of the input is:',XT.shape
print 'size of the subunit is:',sNet.params['sub_r']
print 'subunit location is:',sNet.params['sub']
# plot the loss history
plt.plot(stats['loss_history'])
plt.xlabel('iteration')
plt.ylabel('training loss')
plt.title('Training Loss history')
plt.show()
```

### 1.2 Trainning on a more complete dataset

```python
num_train  = 80;
X = np.reshape(Train_rawimg,(Train_rawimg.shape[0],-1))
y = -Train_target.reshape((Train_target.shape[0],1))
y = y/max(y) # must normalize!!
# split the data into the trainning set and validation set
from sklearn.model_selection import train_test_split
train_ind, test_ind = train_test_split(range(num_train),test_size = 0.1,random_state = 1)
X_train = X[train_ind,:]
X_test = X[test_ind,:]
y_train = y[train_ind]
y_test = y[test_ind]
print X_train.shape,y_train.shape
D = Train_rawimg.shape[1]
num_subunits = 16
G_filter = np.ones((num_subunits,1)) # a fake Gaussian, 
```

```python
sNet = ScratchNet(D,num_subunits,G_filter)
sNet.gridInitilization()
stats = sNet.train(X_train, y_train, X_test, y_test,
            learning_rate=0.001, reg=1e-6,
            num_iters=2000, batch_size = 16,learning_rate_decay=0.9,verbose=True)

print 'Final training loss: ', stats['loss_history'][-1]
print 'shape of the input is:',XT.shape
print 'size of the subunit is:',sNet.params['sub_r']
print 'subunit location is:',sNet.params['sub']
# plot the loss history
plt.plot(stats['loss_history'])
plt.xlabel('iteration')
plt.ylabel('training loss')
plt.title('Training Loss history')
plt.show()
```

```python
plt.plot(stats['train_mse_history'], label='train',color = 'blue')
plt.plot(stats['validation_mse_history'], label='val',color = 'red')
plt.title('Classification accuracy history')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.show()
```

```python
from sklearn.metrics import r2_score
print "r2 score for the validation set:",r2_score(y_test,sNet.predict(X_test))
print "r2 score for the training set:",r2_score(y_train,sNet.predict(X_train))
```

```python
plt.plot(y_train, sNet.predict(X_train),'o')
```
